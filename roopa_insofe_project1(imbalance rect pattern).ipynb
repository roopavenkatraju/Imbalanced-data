Imbalance dataset by keeping Class-1 as a minority class and other two classes will remain the same

$Lets\;consider\;below\;equation\;to\;calculating\;the\;number\;of\;random\;data\;points\;for\;making\;class \;imbalance$ 

$\dfrac{x}{x+n}=y$

$x=y(x+n)$

$x=xy+ny$

$x-xy=ny $

$x(1-y)=ny$

$x=\dfrac{n*y}{1-y}$                        $\;\;\;\;\text{ 'n' value should be number of data points }$

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; $                        $\;\;\;\;\text{ 'y' is the dataset removal percentage }$


$\text{ Example :-} $

$\;\;\;\text{1 % of data from Class 0 has to be considered based on other two classes} $ 

$\;\;\;\text{ lets consider n=2000 and y= 1%}$

$\;\;\;\text{Formula is}$ $x=\dfrac{n*y}{1-y}$ 

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{x=}$  $\dfrac{2000*\frac{1}{100}} {1 - \frac{1}{100}} $

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{x=}$  $\dfrac{2000}{99} $

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{x=20.20}$  $\;\;\;\;\text{ hence 20 random data points should be consider as Class 0 in the dataset.}$

```python
for i in range(1,11):
    y = i/100
    c = ((len(class_2_imp)+len(class_3_imp)) * y) / (1-y)
    z = round(c)
    print(f'{i}% will have {z} samples \n')
```

### Building all supervised models

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,precision_score, f1_score, recall_score, accuracy_score,classification_report,precision_recall_fscore_support
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
def all_model_report(model_names,reports):
    report_data = []
    for i,report in enumerate(reports):
        lines = report.split('\n')
        for line in lines[2:5]:
            row = {}
            row_data = line.split('      ')
            row['Model'] = model_names[i]
            row['class'] ="Class"+row_data[1]
            row['precision'] = float(row_data[2])
            row['recall'] = float(row_data[3])
            row['f1_score'] = float(row_data[4])
            row['support'] = float(row_data[5])
            report_data.append(row)
    return report_data
def split(data_set):
    X = data_set.drop('Target',axis = 1)
    y = data_set['Target']
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,stratify = y,random_state=0)
    models = [['Logistic Regression : \n\n', LogisticRegression(random_state = 0, solver = 'lbfgs',multi_class = 'multinomial')],
              ['Gaussian Naive Bayes : \n\n', GaussianNB()],
              ['Decision_Tree_Classifier : \n\n,',DecisionTreeClassifier(random_state = 0)],
              ['Random_Forest : \n\n',RandomForestClassifier(random_state = 0)],
              ['Knn : \n\n',KNeighborsClassifier(n_neighbors = 3)],
              ['Xg_boost : \n\n',XGBClassifier(random_state = 0)],
              ['Ada_Boost : \n\n',AdaBoostClassifier(random_state = 0)],
              ['Gradient Boosting Classification : \n\n', GradientBoostingClassifier(random_state = 0)],
              ['Bagging Classifier : \n\n',BaggingClassifier(random_state = 0)]]
    report = []
    for name,model in models:
        model = model
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        class_report = classification_report(y_test,y_pred)
        report.append(class_report)       
    return report
```

```python
model_names = ['Logistic Regression',
               'Gaussian Naive Bayes',
               'Decision_Tree_Classifier',
               'Random_Forest',
               'Knn',
               'Xg_boost',
               'Ada_Boost',
               'Gradient Boosting Classification',
               'Bagging Classifier']
```

```python
data = [new_class_1_1,new_class_1_2,new_class_1_3,new_class_1_4,new_class_1_5,new_class_1_6,new_class_1_7,new_class_1_8,new_class_1_9,new_class_1_10]
s_name = "Class_1_"
writer = pd.ExcelWriter('Model_Reports2/Imbalance_Class_1_Models_Reports.xlsx',engine = 'xlsxwriter')
for i in range(len(data)):
    z = split(data[i])
    class_1_models = all_model_report(model_names,z)
    class_1_models = pd.DataFrame.from_dict(class_1_models)
    class_1_models.to_excel(writer,sheet_name = s_name + str(i+1) + '%')
writer.save()
```
